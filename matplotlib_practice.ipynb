```python
# Import dependencies 
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis= ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare is US dollar accumulated
y_axis = [ 10.02, 23.24,39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create Plot
plt.plot(x_axis,y_axis)
```

```python
#Create the plot with ax.plot()
fig, ax=plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#create the plot with ax.plt()
fig= plt.figure()
ax= fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
#creat the plot with ax.plt()
ax=plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Creat the plot and add a label for the legend
plt.plot(x_axis,y_axis,marker="o", color= "pink", linewidth=2, label="Boston")
#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("PyBer Fare by Month")
#Add the legend
plt.legend()

```

```python
#Create the bar graph plot
plt.bar(x_axis,y_axis)
```

```python
#create the plot
plt.bar(x_axis,y_axis, color="pink", label = "Boston")
# Create lables for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create a title
plt.title( "PyBer Fare by Month")
#Add.legend
plt.legend()
```

```python
#create the plot horizontal bar chart
plt.barh(x_axis, y_axis)
```

```python
#create the plot horizontal bar chart
plt.barh(y_axis, x_axis)
```

```python
# invert the y-axis of the previous chart to have "January" at the 
#top and "December" at the bottom
#To invert the y-axis to have the months in ascending order, use the 
#gca() method. The gca() method means "get current axes." We can 
#chain the gca() method to the invert_yaxis() method
plt.barh(x_axis,y_axis, color="yellow", label= "Boston")
plt.title("PyBer Fare by Mont")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
plt.gca().invert_yaxis()
```
